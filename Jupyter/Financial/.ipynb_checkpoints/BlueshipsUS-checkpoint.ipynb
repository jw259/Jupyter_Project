```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import metrics

threshold = 0.15

drop_data = returns_data['MCD'].dropna()

reindex_data = drop_data.reset_index()

del reindex_data['Date']

reindex_data.loc[reindex_data['MCD'] >= threshold, 'CLASS'] = 1
reindex_data.loc[reindex_data['MCD'] < threshold, 'CLASS'] = 0

X = reindex_data[['MCD']].values
y = reindex_data['CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)


# Decision tree

tree = DecisionTreeClassifier(criterion="entropy", max_depth = None)
tree
tree.fit(X_train,y_train)

yhat = tree.predict(X_test)

print("Decision Tree Accuracy: ", metrics.accuracy_score(y_test, yhat))
print("Decision Tree MSE: ", metrics.mean_squared_error(y_test, yhat))
print("Decision Tree F1 Score: ", metrics.f1_score(y_test, yhat, average='weighted'))
print("Decision Tree Jaccard Similarity Score: ", metrics.jaccard_similarity_score(y_test, yhat))


# K Nearest Neighbor

k = 16
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)

yhat = neigh.predict(X_test)

print("kNN Accuracy: ", metrics.accuracy_score(y_test, yhat))
print("kNN MSE: ", metrics.mean_squared_error(y_test, yhat))
print("kNN F1 Score: ", metrics.f1_score(y_test, yhat, average='weighted'))
print("kNN Jaccard Similarity Score: ", metrics.jaccard_similarity_score(y_test, yhat))

# Finding and plotting the best value for k

Ks = 1
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];

for n in range(1,Ks):  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()

# Logistic Regression

log = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)

yhat = log.predict(X_test)

print("Logistic Regression Accuracy: ", metrics.accuracy_score(y_test, yhat))
print("Logistic Regression MSE: ", metrics.mean_squared_error(y_test, yhat))
print("Logistic Regression F1 Score: ", metrics.f1_score(y_test, yhat, average='weighted'))
print("Logistic Regression Jaccard Similarity Score: ", metrics.jaccard_similarity_score(y_test, yhat))
print("Logistic Regression Log Loss: ", metrics.log_loss(y_test, yhat))

# Support Vector Machine

clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train)
yhat = clf.predict(X_test)

print("SVM Accuracy: ", metrics.accuracy_score(y_test, yhat))
print("SVM MSE: ", metrics.mean_squared_error(y_test, yhat))
print("SVM F1 Score: ", metrics.f1_score(y_test, yhat, average='weighted'))
print("SVM Jaccard Similarity Score: ", metrics.jaccard_similarity_score(y_test, yhat))
```
