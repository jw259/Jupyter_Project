```python
import pandas as pd
from pandas_datareader import data
import numpy as np
import seaborn as sns
from datetime import date
```

```python
tickers = ['FB', 'AAPL', 'AMZN', 'NFLX', 'GOOG', 'MSFT', 'QCOM', 'NVDA', 'ZM']

start_date = '1980-12-12'
end_date = date.today()

financial_data = data.DataReader(tickers, 'yahoo', start_date, end_date)
```

```python
financial_data.head(-1)
```

```python
close_data = financial_data['Close']
close_data.head(-1)
```

```python
returns_data = close_data.pct_change()
returns_data.head(-1)
```

```python
mean = returns_data.mean() * 100
print(mean)
```

```python
std = returns_data.std() * 100
print(std)
```

```python
returns_data.describe()
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import metrics

threshold = 0.001

drop_data = returns_data.dropna()

print(drop_data)

reindex_data = drop_data.reset_index()

del reindex_data['Date']

print(reindex_data)

reindex_data.loc[reindex_data['AAPL'] >= threshold, 'CLASS'] = 1
reindex_data.loc[reindex_data['AAPL'] < threshold, 'CLASS'] = 0

print(reindex_data)

x = reindex_data['AAPL']
y = reindex_data['CLASS']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)

print(x_train)
print(y_train)
print(x_test)
print(y_test)

clf =svm.SVC(kernel='rbf')
clf.fit(x_train.values.reshape(-1, 1), y_train)

yhat = clf.predict(x_test)

#print("SVM Accuracy: ", metrics.accuracy_score(y_test, yhat))
#print("SVM MSE: ", metrics.mean_squared_error(y_test, yhat))
#print("SVM F1 Score: ", metrics.f1_score(y_test, yhat, average='weighted'))
#print("SVM Jaccard Similarity Score: ", metrics.jaccard_similarity_score(y_test, yhat))
```
